```python
def von_neumann_entropy(rho):
    import numpy as np
    from scipy import linalg as la
    R = rho*(la.logm(rho)/la.logm(np.matrix([[2]]))) #dividing loga/logb since the calculation 
                                                     #is wrt qubits ie base 2 system
    S = -np.matrix.trace(R)
    return(S)
```

```python
import numpy as np
# Define spin-up as convention
u = np.matrix([[1],
               [0]])

# Define spin-down as convention
d = np.matrix([[0],
               [1]])

# Define the pure state |0><0|
P11 = np.dot(u, u.H)

# Define the pure state |1><1|
P22 = np.dot(d, d.H)
```

```python
import matplotlib.pyplot as plt

p = np.arange(0.001, 1., 0.01)

# create list of von Neumann entropies
vn_y = list()
for value in p:
    rho_p = value*P11 + (1-value)*(P22)
    vn_y.append(von_neumann_entropy(rho_p))

# matrices->numbers for ease of plotting
y = list()
for value in vn_y:
    y.append(value.tolist()[0][0])

# plot p vs. S(rho_p)
plt.xlabel('p')
plt.ylabel('von Neumann Entropy')
plt.title('S(rho_p)')
plt.plot(p, y, color='tab:green')
plt.show()
p = np.arange(0.001, 1., 0.01)

#shannon entropy for the same system
plt.xlabel('Probability')
plt.ylabel('Shannon Entropy')
plt.title('H(p, 1-p)')
plt.plot(p, -( p*np.log2(p) + (1-p)*np.log2(1-p)), 'b.')
plt.show()
```
